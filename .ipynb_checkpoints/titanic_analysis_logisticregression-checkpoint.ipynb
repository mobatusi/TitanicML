## Fit A Logistic Regression

A logistic regression is a binary classifier (i.e. the target vector can only take two values). In a logistic regression, a linear model (e.g. Î²0+Î²1) is included in a logistic also called sigmoid function, 
\begin{equation*}
\frac{1}{(1+e-z)}
\end{equation*}

such that: 

\begin{equation*}
P(yi=1âˆ£X) = \frac{1}{1+e^âˆ’(Î²0+Î²1x)} 
\end{equation*}

where P(yi=1âˆ£X) is the probability of the ith observationâ€™s target value, yi, being class 1, X is the training data, Î²0 and Î²1 are the parameters to be learned, and e is Eulerâ€™s number.

### Use Sex, Age, SibSp, Parch, and Embarked as training data

```python
# Load data with only certain features
X = df_train[['Pclass', 'Sex', 'Age','SibSp','Parch', 'Embarked']][df_train.Age.isna() == False]
y = df_train.Survived[df_train.Age.isna() == False]
```

### Convert categorial data to numerical code

### Sex
male = 1
female = 0

### Embarked
C = 0

```python
clean up null alues
```

```python
X["Embarked"].value_counts()
```

For the sake of simplicity, just fill in the value with the S (since that is the most common value)

```python
X = X.fillna({"Embarked": "S"})
```

It worked

```python
X[X.isnull().any(axis=1)]
```

```python
X.dtypes
```

Convert the objects to category to encode

```python
X["Sex"] = X["Sex"].astype('category')
X["Embarked"] = X["Embarked"].astype('category')
X.dtypes
```

```python
X["Sex_cat"] = X["Sex"].cat.codes
X["Embarked_cat"] = X["Embarked"].cat.codes
X.head()
```

```python
X_encoded = X[['Pclass', 'Sex_cat', 'Age','SibSp','Parch', 'Embarked_cat']]
```

### Standardize Features

```python
scaler = StandardScaler()
X_std = scaler.fit_transform(X_encoded)
```

### Create logistic regression object

```python
clf = LogisticRegression(random_state=0)
```

```python
clf
```

### Train Logistic Regression

```python
# Train model
model = clf.fit(X_std, y)
```

### Create Pipeline

```python
# Create standardizer
standardizer = StandardScaler()

# Create logistic regression
logit = LogisticRegression()

# Create a pipeline that standardizes, then runs logistic regression
pipeline = make_pipeline(standardizer, logit)
```

### Create k-Fold Cross-Validation

```python
# Create k-Fold cross-validation
kf = KFold(n_splits=10, shuffle=True, random_state=1)
```

### Conduct k-Fold Cross-Validation

```python
# Do k-fold cross-validation
cv_results = cross_val_score(pipeline, # Pipeline
                             X_std, # Feature matrix
                             y, # Target vector
                             cv=kf, # Cross-validation technique
                             scoring="accuracy", # Loss function
                             n_jobs=-1) # Use all CPU scores
```

```python
cv_results
```

### Calculate Mean Performance Score

```python
# Calculate mean
cv_results.mean()
```
